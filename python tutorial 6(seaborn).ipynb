```python
sns.barplot(x='preferred_foot',y='weak_foot',data=fifa)
plt.show()
```

```python
sns.barplot(x='preferred_foot',y='weak_foot',data=fifa,palette='rocket')
plt.show()
```

```python
#seaborn histogram/displot----distribution plot/histplot----histogram
```

```python
sns.distplot(fifa['overall'])  #dist
plt.show()
```

```python
sns.displot(fifa['overall'])
plt.show()
```

```python
sns.histplot(fifa['overall'])
plt.show()
```

```python
sns.distplot(fifa['overall'],color='green')  #dist
plt.show()
```

```python
sns.distplot(fifa['overall'],color='green',hist=False)  #dist
plt.show()
```

```python
sns.distplot(fifa['overall'],color='green',kde=False)  #dist
plt.show()
```

```python
sns.distplot(fifa['overall'],color='red',kde=False,bins=100)
plt.show()
```

```python
sns.distplot(fifa['overall'],color='red',kde=False,bins=100,vertical=True)
plt.show()
```

```python
#jotntplot
```

```python
sns.jointplot(x='overall',y='weak_foot',data=fifa,color='olive')
plt.show()
```

```python
sns.jointplot(x='overall',y='weak_foot',data=fifa,color='olive',kind='reg')#regression
plt.show()
```
