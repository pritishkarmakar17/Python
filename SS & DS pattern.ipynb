```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
from scipy.fft import fft, fftfreq
```

```python
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 100
```

```python
x=[0]*20000+[0.6]*20+[0]*10+[1]*20+[0]*20000
#x=[0]*2000+[1]*20+[0]*2000
t=np.linspace(0,10,len(x))

plt.plot(t,x)
plt.show()
```

```python
N = len(x)
yf = fft(x)
f = fftfreq(N,np.diff(t)[0])
```

```python
plt.scatter(f,np.abs(yf),s=0.2)
```
