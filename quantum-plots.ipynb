```python
E=lambda x,m,y,n: (2/12)*np.sin(m*np.pi*x/12)*np.sin(n*np.pi*y/12)
m1=int(input('n1 :'))
m2=int(input('n2 :'))
a1=np.linspace(0,12,500)
xv,yv=np.meshgrid(a1,a1)
zv=E(xv,m1,yv,m2)
plt.contourf(xv,yv,zv,levels=300,cmap='viridis')
plt.colorbar()
plt.title('$\psi$ for particle in a 2d box')
```

```python

A=1
n1= int(input('n : '))
psi= lambda x,n: A*(1/(2**n*factorial(n))*special.hermite(n, monic='True')(x)*np.exp(-x**2/2))
X=np.linspace(-10,10,5000)
plt.xlabel('x')
plt.ylabel('$\psi^2$')
plt.title('Harmonic oscillator')
N=1/(max(psi(X,n1)**2)-min(psi(X,n1)**2))
plt.plot(X,N*psi(X,n1)**2)
```

```python
def psi2plot(n1):
    A=1
    B=1
    psi= lambda x,n: A*(1/(2**n*factorial(n))*special.hermite(n, monic='True')(x)*np.exp(-x**2/2))
    X=np.linspace(-10,10,5000)
    plt.xlabel('x')
    plt.ylabel('$\psi^2$')
    plt.title('Harmonic oscillator')
    N=1/(max(psi(X,n1)**2)-min(psi(X,n1)**2))
    plt.plot(X,0.9*N*psi(B*X,n1)**2+n1)
```

```python
k=int(input())
for i in range(k):
    psi2plot(i)
plt.show()
```
