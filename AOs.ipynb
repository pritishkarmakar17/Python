```python
import numpy as np
import matplotlib.pyplot as plt
import math as m
```

```python
R=lambda r,Y: (3.5*r-r**2)*np.exp(-r/2)*Y
a1=np.linspace(-12,12,500)
xv,yv=np.meshgrid(a1,a1)
zv=R(np.sqrt(xv**2+yv**2),xv/np.sqrt(xv**2+yv**2))
plt.contourf(xv,yv,zv,levels=30,cmap='bwr')
plt.colorbar()
```

```python
zv=R(np.sqrt(xv**2+yv**2),xv*yv/(xv**2+yv**2))
plt.contourf(xv,yv,zv,levels=30,cmap='bwr')
plt.colorbar()
```
