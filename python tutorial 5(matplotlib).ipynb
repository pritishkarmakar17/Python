```python
#line plot
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
x
```

```python
y=x**2
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title('line plot')
plt.xlabel('x value')
plt.ylabel('y value')
plt.show()
```

```python
plt.plot(x,y,color='r',linestyle=':',linewidth=2)
plt.show()
```

```python
x=np.arange(0,6,0.5)
y1=2*x
y2=x**2
```

```python
plt.plot(x,y1,color='g',linestyle='-',linewidth=2)
plt.plot(x,y2,color='r',linestyle='-.',linewidth=2)
plt.title('line plot')
plt.xlabel('x value')
plt.ylabel('y value')
plt.grid(True)
plt.show()
```

```python
x=np.arange(0,11)
y1=2*x
y2=x**2
```

```python
plt.subplot(1,2,1) #(row,column,index of plot)
plt.plot(x,y1,color='g',linestyle='-',linewidth=2)
plt.title('linear plot')
plt.xlabel('x value')
plt.ylabel('y value')
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(x,y2,color='r',linestyle='-.',linewidth=2)
plt.title('exponential plot')
plt.xlabel('x value')
plt.ylabel('y value')
plt.grid(True)

plt.show()
```

```python
plt.plot(x,y1,color='g',linestyle='-',linewidth=2)
plt.title('linear plot')
plt.xlabel('x value')
plt.ylabel('y value')
plt.grid(True)

plt.subplot(2,1,2)
plt.plot(x,y2,color='r',linestyle='-.',linewidth=2)
plt.title('exponential plot')
plt.xlabel('x value')
plt.ylabel('y value')
plt.grid(True)

plt.show()
```

```python
#bar plot
```

```python
footballers={'ronaldo':46,'messi':40,'lewandowski':54}
```

```python
name=list(footballers.keys())
goals=list(footballers.values())
```

```python
plt.bar(name,goals) #(catagory, values)
plt.title('goals in a season')
plt.xlabel('players')
plt.ylabel('goals')
plt.show()
```

```python
plt.barh(name,goals, color='r') #(catagory, values)
plt.title('goals in a season')
plt.ylabel('players')
plt.xlabel('goals')
plt.grid(True,linestyle=':')
plt.show()
```

```python
#scatter plot
x=[1,3,3,4,4,6,7,7,9,9,10]
y=[0,-2,2,-4,4,0,-3,3,-4,4,0]
plt.scatter(x,y,marker='*',c='r',s=100)
plt.title('4-3-3')
plt.ylabel('y')
plt.xlabel('x')
plt.grid(True,linestyle=':')
plt.show()
```

```python
x=[1,3,3,4,4,6,7,7,9,9,10]
y=[0,-2,2,-4,4,0,-3,3,-4,4,0]
plt.scatter(x,y,marker='.',c='g',s=500)  #c-color s-size
plt.title('4-3-3')
plt.ylabel('y')
plt.xlabel('x')
plt.grid(True,linestyle=':')
plt.show()
```

```python
x=[10,20,30,40,50,60,70,80,90]
a=[2,4,5,2,7,8,9,5,9]
b=[4,5,6,3,8,0,2,6,8]
plt.scatter(x,a,marker='.',c='b',s=100)
plt.scatter(x,b,marker='.',c='r',s=100)
plt.ylabel('y')
plt.xlabel('x')
plt.grid(True,linestyle=':',linewidth=1)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(x,a,color='g')
plt.title('g1')
plt.xlabel('x value')
plt.ylabel('y value')
plt.grid(True,color='y')

plt.subplot(1,2,2)
plt.scatter(x,b,color='r')
plt.title('g2')
plt.xlabel('x value')
plt.ylabel('y value')
plt.grid(True,color='c')

plt.show()
```
